```python
def convert_age_range_to_numeric(age_range):
    if "or older" in age_range:
        return 80
    elif "-" in age_range:
        start, end = age_range.split("-")
        return (int(start) + int(end)) / 2
    else:
        return None
```

```python
numeric_ages = pd.to_numeric([convert_age_range_to_numeric(age) for age in df['AgeCategory']], errors='coerce')
```

```python
df['AgeCategory'] = numeric_ages #Age category as turned into float datatype
```
